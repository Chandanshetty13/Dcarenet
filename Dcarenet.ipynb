```python
#############################################################
# SINGLE-CODE-BLOCK PIPELINE WITH FIXED GENERATOR OUTPUT
# GAN Augmentation + Preprocessing + ResNet-CapsNet Ensemble + 
# Hypothetical Cataract/Glaucoma Anomaly Detection
#
# EXTENDED VERSION: 
#  1) Model Saving
#  2) Separate Performance Metrics
#  3) Comparative Analysis
#  4) Visualizations (bar, line, confusion, ROC-AUC, feature maps)
#############################################################

import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# TensorFlow / Keras
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50

# For saving/loading scikit-learn models
import pickle
#from joblib import dump, load  # alternative approach

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, confusion_matrix, classification_report, 
                             roc_curve, roc_auc_score)
import seaborn as sns

# For reproducibility (optional)
np.random.seed(42)
tf.random.set_seed(42)

# =========================
# 1. DATA PREPROCESSING
# =========================
def preprocess_image(img_path, target_size=224):
    """
    Basic function: loads an image, optionally crops/adjusts,
    then resizes to (target_size x target_size), and normalizes.
    """
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # [Optional] Crop black borders / enhance contrast if needed
    img = cv2.resize(img, (target_size, target_size))
    img = img.astype(np.float32) / 255.0
    return img

df = pd.read_csv("train.csv")  # Must have 'id_code' and 'diagnosis' columns

train_df, val_df = train_test_split(
    df, test_size=0.2, stratify=df["diagnosis"], random_state=42
)

import math

class DRDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, df, img_dir, batch_size=16, shuffle=True, augment=False):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.augment = augment
        self.on_epoch_end()

    def __len__(self):
        return math.ceil(len(self.df) / self.batch_size)

    @property
    def num_batches(self):
        return self.__len__()

    def on_epoch_end(self):
        self.indexes = np.arange(len(self.df))
        if self.shuffle:
            np.random.shuffle(self.indexes)

    def __getitem__(self, idx):
        batch_indexes = self.indexes[idx*self.batch_size : (idx+1)*self.batch_size]
        images, labels = [], []
        for i in batch_indexes:
            row = self.df.iloc[i]
            img_path = os.path.join(self.img_dir, f"{row['id_code']}.png")
            img = preprocess_image(img_path, 224)
            if self.augment and np.random.rand() < 0.5:
                img = np.fliplr(img)
            images.append(img)
            labels.append(row["diagnosis"])
        return np.array(images), np.array(labels)

train_generator = DRDataGenerator( 
    df=train_df,
    img_dir="train_images",  # folder with .png images
    batch_size=16,
    shuffle=True,
    augment=True
)

val_generator = DRDataGenerator(
    df=val_df,
    img_dir="train_images",
    batch_size=16,
    shuffle=False,
    augment=False
)

NUM_CLASSES = 5  # DR labels: 0,1,2,3,4

# =========================
# 2. cGAN for Data Augmentation (Fixed Generator Output)
# =========================
LATENT_DIM = 100

def build_generator():
    noise_input = layers.Input(shape=(LATENT_DIM,))
    label_input = layers.Input(shape=(1,))
    
    label_embed = layers.Embedding(NUM_CLASSES, 50)(label_input)
    label_embed = layers.Flatten()(label_embed)

    combined = layers.Concatenate()([noise_input, label_embed])
    # Produce 14x14x256 so we can do 4 upsamplings to reach 224×224
    x = layers.Dense(14 * 14 * 256, activation='relu')(combined)
    x = layers.Reshape((14, 14, 256))(x)
    x = layers.Conv2DTranspose(128, 4, strides=2, padding='same', activation='relu')(x)   # 28×28
    x = layers.Conv2DTranspose(64, 4, strides=2, padding='same', activation='relu')(x)    # 56×56
    x = layers.Conv2DTranspose(32, 4, strides=2, padding='same', activation='relu')(x)    # 112×112
    out = layers.Conv2DTranspose(3, 4, strides=2, padding='same', activation='tanh')(x)   # 224×224×3

    generator = models.Model([noise_input, label_input], out, name="Generator")
    return generator

def build_discriminator():
    img_input = layers.Input(shape=(224,224,3))
    label_input = layers.Input(shape=(1,))

    label_embed = layers.Embedding(NUM_CLASSES, 50)(label_input)
    label_embed = layers.Flatten()(label_embed)
    label_embed = layers.Dense(224*224, activation='relu')(label_embed)
    label_embed = layers.Reshape((224,224,1))(label_embed)

    combined_input = layers.Concatenate(axis=-1)([img_input, label_embed])
    x = layers.Conv2D(64, 4, strides=2, padding='same')(combined_input)  # 112×112
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Conv2D(128, 4, strides=2, padding='same')(x)             # 56×56
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Flatten()(x)
    out = layers.Dense(1, activation='sigmoid')(x)

    discriminator = models.Model([img_input, label_input], out, name="Discriminator")
    return discriminator

generator = build_generator()
discriminator = build_discriminator()

cross_entropy = tf.keras.losses.BinaryCrossentropy()
g_optimizer = tf.keras.optimizers.Adam(1e-4)
d_optimizer = tf.keras.optimizers.Adam(1e-4)

MINORITY_CLASS = 4
BATCH_SIZE_GAN = 16

def get_real_batch(df, img_dir, minority_class=4, batch_size=16):
    df_minority = df[df["diagnosis"] == minority_class].sample(n=batch_size, replace=True)
    real_imgs, real_labels = [], []
    for i in range(len(df_minority)):
        row = df_minority.iloc[i]
        img_path = os.path.join(img_dir, f"{row['id_code']}.png")
        img = preprocess_image(img_path, target_size=224)
        real_imgs.append(img)
        real_labels.append(minority_class)
    return np.array(real_imgs), np.array(real_labels).reshape(-1,1)

@tf.function
def train_step_gan(real_imgs, real_labels):
    # 1. Train Discriminator
    noise = tf.random.normal([BATCH_SIZE_GAN, LATENT_DIM])
    gen_labels = tf.random.uniform([BATCH_SIZE_GAN, 1], minval=0, maxval=NUM_CLASSES, dtype=tf.int32)
    fake_imgs = generator([noise, gen_labels], training=True)

    with tf.GradientTape() as tape_d:
        pred_real = discriminator([real_imgs, real_labels], training=True)
        pred_fake = discriminator([fake_imgs, gen_labels], training=True)
        d_loss_real = cross_entropy(tf.ones_like(pred_real), pred_real)
        d_loss_fake = cross_entropy(tf.zeros_like(pred_fake), pred_fake)
        d_loss = (d_loss_real + d_loss_fake) / 2
    grads_d = tape_d.gradient(d_loss, discriminator.trainable_variables)
    d_optimizer.apply_gradients(zip(grads_d, discriminator.trainable_variables))

    # 2. Train Generator
    noise = tf.random.normal([BATCH_SIZE_GAN, LATENT_DIM])
    gen_labels = tf.random.uniform([BATCH_SIZE_GAN, 1], minval=0, maxval=NUM_CLASSES, dtype=tf.int32)
    with tf.GradientTape() as tape_g:
        gen_imgs = generator([noise, gen_labels], training=True)
        pred_fake = discriminator([gen_imgs, gen_labels], training=True)
        g_loss = cross_entropy(tf.ones_like(pred_fake), pred_fake)
    grads_g = tape_g.gradient(g_loss, generator.trainable_variables)
    g_optimizer.apply_gradients(zip(grads_g, generator.trainable_variables))

    return d_loss, g_loss

EPOCHS_GAN = 200  # Demo: you can increase
for epoch in range(EPOCHS_GAN):
    real_imgs, real_labels = get_real_batch(train_df, "train_images", 
                                            minority_class=MINORITY_CLASS,
                                            batch_size=BATCH_SIZE_GAN)
    d_loss, g_loss = train_step_gan(real_imgs, real_labels)
    if epoch % 5 == 0:
        print(f"[GAN Epoch {epoch}] d_loss: {d_loss.numpy():.4f}, g_loss: {g_loss.numpy():.4f}")

# Generate synthetic 224×224 samples for the minority class
NUM_SYNTHETIC = 10
noise = tf.random.normal([NUM_SYNTHETIC, LATENT_DIM])
fake_labels = tf.ones([NUM_SYNTHETIC, 1], dtype=tf.int32) * MINORITY_CLASS
synthetic_imgs = generator.predict([noise, fake_labels])
synthetic_imgs = (synthetic_imgs + 1.0) / 2.0  # scale from tanh [-1,1] to [0,1]

# =========================
# 3. TRAIN MODELS (ResNet + CapsNet)
# =========================
base_resnet = ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))
for layer in base_resnet.layers:
    layer.trainable = False

x = layers.GlobalAveragePooling2D()(base_resnet.output)
x = layers.Dense(256, activation='relu')(x)
out = layers.Dense(NUM_CLASSES, activation='softmax')(x)

resnet_model = models.Model(base_resnet.input, out)
resnet_model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

print("Training ResNet Model...")
EPOCHS_RESNET = 100  # For demonstration
history_resnet = resnet_model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS_RESNET
)

# CapsNet building blocks
def squash(vectors, axis=-1):
    s_squared_norm = tf.reduce_sum(tf.square(vectors), axis, keepdims=True)
    scale = s_squared_norm / (1 + s_squared_norm) / tf.sqrt(s_squared_norm + 1e-9)
    return scale * vectors

class CapsuleLayer(layers.Layer):
    def __init__(self, num_capsules, dim_capsules, routings=3, **kwargs):
        super(CapsuleLayer, self).__init__(**kwargs)
        self.num_capsules = num_capsules
        self.dim_capsules = dim_capsules
        self.routings = routings

    def build(self, input_shape):
        self.input_dim = input_shape[-1]
        self.kernel = self.add_weight(
            shape=(self.input_dim, self.num_capsules * self.dim_capsules),
            initializer='glorot_uniform',
            name='capsule_kernel'
        )

    def call(self, inputs):
        # inputs: [batch_size, ..., input_dim]
        inputs_reshaped = tf.reshape(inputs, [tf.shape(inputs)[0], -1, self.input_dim])
        # u_hat: [batch_size, num_vecs, num_capsules * dim_capsules]
        u_hat = tf.einsum('bij,jk->bik', inputs_reshaped, self.kernel)
        # reshape to [batch_size, num_vecs, num_capsules, dim_capsules]
        u_hat_reshaped = tf.reshape(u_hat, [tf.shape(u_hat)[0], -1, self.num_capsules, self.dim_capsules])
        # Simplified routing: average
        outputs = tf.reduce_mean(u_hat_reshaped, axis=1)  # [batch_size, num_capsules, dim_capsules]
        outputs = squash(outputs)
        return outputs

def build_capsnet(input_shape=(224, 224, 3), n_class=5):
    x_input = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, (3, 3), strides=2, padding='same', activation='relu')(x_input)  # 112
    x = layers.Conv2D(64, (3, 3), strides=2, padding='same', activation='relu')(x)        # 56
    x = layers.Reshape((-1, 8))(x)  # Reshape for capsule input
    x = layers.Lambda(squash)(x)

    # Capsule layer
    class_caps = CapsuleLayer(num_capsules=n_class, dim_capsules=16, routings=3)(x)
    out_caps = layers.Lambda(lambda z: tf.sqrt(tf.reduce_sum(tf.square(z), axis=2)))(class_caps)

    model = models.Model(x_input, out_caps)
    return model

capsnet_model = build_capsnet()
capsnet_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print("Training CapsNet Model...")
EPOCHS_CAPS = 100
history_capsnet = capsnet_model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS_CAPS
)

# =========================
# 4. ENSEMBLING VIA STACKING
# =========================
print("Generating validation set predictions for stacking...")

val_imgs = []
val_labels = []
for i in range(len(val_df)):
    row = val_df.iloc[i]
    img_path = os.path.join("train_images", f"{row['id_code']}.png")
    img = preprocess_image(img_path, 224)
    val_imgs.append(img)
    val_labels.append(row['diagnosis'])

val_imgs = np.array(val_imgs)
val_labels = np.array(val_labels)

resnet_preds = resnet_model.predict(val_imgs)   # shape [N,5]
capsnet_preds = capsnet_model.predict(val_imgs) # shape [N,5]

# Probability vectors combined
X_stack = np.concatenate([resnet_preds, capsnet_preds], axis=1)  # shape [N,10]
y_stack = val_labels

# Logistic Regression as meta-learner
meta_model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
meta_model.fit(X_stack, y_stack)

y_pred_stack = meta_model.predict(X_stack)
stack_accuracy = accuracy_score(y_stack, y_pred_stack)
print("Stacked Ensemble Accuracy:", stack_accuracy)

# =========================
# 5. HYPOTHETICAL ANOMALY DETECTION (Cataract/Glaucoma)
# =========================
def build_autoencoder(input_shape=(224,224,3)):
    """
    Simple conv autoencoder for anomaly detection.
    Trained on DR=0 images as 'healthy'.
    """
    encoder_input = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(encoder_input)
    x = layers.MaxPooling2D((2,2))(x)  #112
    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2,2))(x)  #56
    x = layers.Flatten()(x)
    bottleneck = layers.Dense(256, activation='relu')(x)
    
    x = layers.Dense(64*56*56, activation='relu')(bottleneck)
    x = layers.Reshape((56,56,64))(x)
    x = layers.UpSampling2D((2,2))(x)    #112
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2,2))(x)    #224
    decoded = layers.Conv2D(3, (3,3), activation='sigmoid', padding='same')(x)
    
    autoencoder = models.Model(encoder_input, decoded)
    return autoencoder

df_no_dr = train_df[train_df["diagnosis"] == 0].copy().reset_index(drop=True)
autoencoder = build_autoencoder()
autoencoder.compile(optimizer='adam', loss='mse')

class NoDRDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, df, img_dir, batch_size=16):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.batch_size = batch_size
        self.indexes = np.arange(len(self.df))
        np.random.shuffle(self.indexes)

    def __len__(self):
        return len(self.df) // self.batch_size

    def on_epoch_end(self):
        np.random.shuffle(self.indexes)

    def __getitem__(self, idx):
        batch_indexes = self.indexes[idx*self.batch_size : (idx+1)*self.batch_size]
        imgs = []
        for i in batch_indexes:
            row = self.df.iloc[i]
            img_path = os.path.join(self.img_dir, f"{row['id_code']}.png")
            img = preprocess_image(img_path, 224)
            imgs.append(img)
        return np.array(imgs), np.array(imgs)  # autoencoder learns to reconstruct

no_dr_gen = NoDRDataGenerator(df_no_dr, "train_images", batch_size=16)
print("Training Autoencoder for anomaly detection...")
autoencoder.fit(no_dr_gen, epochs=200)  # quick demo

def anomaly_score(img, autoencoder):
    inp = np.expand_dims(img, axis=0)
    recon = autoencoder.predict(inp)
    return np.mean((recon - inp)**2)

# Example usage
test_img_path = "train_images/0a4e1a29ffff.png"
if os.path.exists(test_img_path):
    test_img = preprocess_image(test_img_path, 224)
    score = anomaly_score(test_img, autoencoder)
    threshold = 0.02
    if score > threshold:
        print(f"Anomaly score = {score:.4f} > {threshold} --> Possible cataract/glaucoma/other.")
    else:
        print(f"Anomaly score = {score:.4f} <= {threshold} --> Likely normal or DR=0-like.")
else:
    print(f"Test image path '{test_img_path}' not found. Skipping anomaly detection example.")

###########################################################
# 6. SAVING ALL MODELS
###########################################################
# Keras models can be saved as .h5 or SavedModel format
generator.save("generator_model.h5")
discriminator.save("discriminator_model.h5")
resnet_model.save("resnet_model.h5")
capsnet_model.save("capsnet_model.h5")
autoencoder.save("autoencoder_model.h5")

# Logistic Regression (scikit-learn) - can use pickle or joblib
with open("meta_model.pkl", "wb") as f:
    pickle.dump(meta_model, f)

print("\nModels saved successfully!\n")

###########################################################
# 7. PERFORMANCE METRICS & COMPARATIVE ANALYSIS
#    (ResNet, CapsNet, Ensemble)
###########################################################
# For demonstration, we measure performance on the same validation set (val_imgs, val_labels).

# --- 7.1 Evaluate ResNet ---
resnet_eval = resnet_model.evaluate(val_imgs, val_labels, verbose=0)
resnet_loss, resnet_acc = resnet_eval[0], resnet_eval[1]
print("ResNet - Loss: {:.4f}, Accuracy: {:.4f}".format(resnet_loss, resnet_acc))

resnet_y_pred_prob = resnet_model.predict(val_imgs)  # shape [N, 5]
resnet_y_pred = np.argmax(resnet_y_pred_prob, axis=1)

# Confusion matrix
cm_resnet = confusion_matrix(val_labels, resnet_y_pred)
print("ResNet Confusion Matrix:")
print(cm_resnet)

print("ResNet Classification Report:")
print(classification_report(val_labels, resnet_y_pred))

# ROC-AUC (multi-class)
# Using 'ovr' approach for multi-class:
try:
    resnet_roc_auc = roc_auc_score(val_labels, resnet_y_pred_prob, multi_class='ovr')
    print("ResNet ROC-AUC (OVR): {:.4f}".format(resnet_roc_auc))
except:
    print("Cannot compute ROC-AUC for ResNet. Possibly only one class in val_labels?")

# --- 7.2 Evaluate CapsNet ---
capsnet_eval = capsnet_model.evaluate(val_imgs, val_labels, verbose=0)
capsnet_loss, capsnet_acc = capsnet_eval[0], capsnet_eval[1]
print("\nCapsNet - Loss: {:.4f}, Accuracy: {:.4f}".format(capsnet_loss, capsnet_acc))

capsnet_y_pred_prob = capsnet_model.predict(val_imgs)  # shape [N, 5]
capsnet_y_pred = np.argmax(capsnet_y_pred_prob, axis=1)

cm_capsnet = confusion_matrix(val_labels, capsnet_y_pred)
print("CapsNet Confusion Matrix:")
print(cm_capsnet)

print("CapsNet Classification Report:")
print(classification_report(val_labels, capsnet_y_pred))

try:
    capsnet_roc_auc = roc_auc_score(val_labels, capsnet_y_pred_prob, multi_class='ovr')
    print("CapsNet ROC-AUC (OVR): {:.4f}".format(capsnet_roc_auc))
except:
    print("Cannot compute ROC-AUC for CapsNet.")

# --- 7.3 Evaluate Ensemble (Stacking) ---
ensemble_y_pred = y_pred_stack  # we already have from meta_model
ensemble_y_pred_prob = meta_model.predict_proba(X_stack)  # shape [N, 5] 
# (Note: If your logistic regression is multi-class, you'll get [N,5] classes.)

ensemble_acc = accuracy_score(val_labels, ensemble_y_pred)
print("\nEnsemble - Accuracy: {:.4f}".format(ensemble_acc))

cm_ensemble = confusion_matrix(val_labels, ensemble_y_pred)
print("Ensemble Confusion Matrix:")
print(cm_ensemble)

print("Ensemble Classification Report:")
print(classification_report(val_labels, ensemble_y_pred))

# For ROC-AUC with sklearn LR in multi-class:
try:
    ensemble_roc_auc = roc_auc_score(val_labels, ensemble_y_pred_prob, multi_class='ovr')
    print("Ensemble ROC-AUC (OVR): {:.4f}".format(ensemble_roc_auc))
except:
    print("Cannot compute ROC-AUC for Ensemble.")

# Collect final metrics for comparison
all_results = {
    "ResNet": {"Loss": resnet_loss, "Accuracy": resnet_acc, "ROC-AUC": (resnet_roc_auc if 'resnet_roc_auc' in locals() else None)},
    "CapsNet": {"Loss": capsnet_loss, "Accuracy": capsnet_acc, "ROC-AUC": (capsnet_roc_auc if 'capsnet_roc_auc' in locals() else None)},
    "Ensemble": {"Accuracy": ensemble_acc, "ROC-AUC": (ensemble_roc_auc if 'ensemble_roc_auc' in locals() else None)},
}

print("\n=== Comparative Analysis ===")
for model_name, metrics_dict in all_results.items():
    print(f"{model_name}: {metrics_dict}")

###########################################################
# 8. VISUALIZATIONS
###########################################################
# 8.1 Training History (Line Plots) for ResNet & CapsNet
plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
plt.plot(history_resnet.history['accuracy'], label='train_acc')
plt.plot(history_resnet.history['val_accuracy'], label='val_acc')
plt.title("ResNet Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history_resnet.history['loss'], label='train_loss')
plt.plot(history_resnet.history['val_loss'], label='val_loss')
plt.title("ResNet Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
plt.plot(history_capsnet.history['accuracy'], label='train_acc')
plt.plot(history_capsnet.history['val_accuracy'], label='val_acc')
plt.title("CapsNet Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history_capsnet.history['loss'], label='train_loss')
plt.plot(history_capsnet.history['val_loss'], label='val_loss')
plt.title("CapsNet Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

# 8.2 Bar Graphs for final Accuracy of each model
model_names = ["ResNet", "CapsNet", "Ensemble"]
accuracy_scores = [
    resnet_acc,
    capsnet_acc,
    ensemble_acc
]

plt.figure(figsize=(6, 4))
bars = plt.bar(model_names, accuracy_scores, color=['skyblue','orange','green'])
plt.title("Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + 0.3, yval + 0.005, f"{yval:.3f}")
plt.ylim([0,1])
plt.show()

# 8.3 Confusion Matrices (heatmaps)
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

plot_confusion_matrix(cm_resnet, "ResNet Confusion Matrix")
plot_confusion_matrix(cm_capsnet, "CapsNet Confusion Matrix")
plot_confusion_matrix(cm_ensemble, "Ensemble Confusion Matrix")

# 8.4 ROC Curves for each model (multi-class one-vs-rest approach)
def plot_multiclass_roc(y_true, y_score, n_classes=5, title=""):
    """
    y_true: shape [N, ] with class indices
    y_score: shape [N, n_classes] with predicted probabilities
    """
    # Binarize the labels
    from sklearn.preprocessing import label_binarize
    y_true_bin = label_binarize(y_true, classes=np.arange(n_classes))

    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_score[:, i])
        roc_auc[i] = roc_auc_score(y_true_bin[:, i], y_score[:, i])

    colors = ['blue','red','green','orange','purple']
    plt.figure()
    for i in range(n_classes):
        plt.plot(fpr[i], tpr[i], color=colors[i % len(colors)],
                 label=f"Class {i} (area = {roc_auc[i]:.2f})")
    plt.plot([0,1],[0,1],'k--')
    plt.xlim([0,1])
    plt.ylim([0,1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.show()

# Plot if feasible
# ResNet
try:
    plot_multiclass_roc(val_labels, resnet_y_pred_prob, n_classes=NUM_CLASSES, 
                        title="ResNet ROC Curves")
except:
    pass

# CapsNet
try:
    plot_multiclass_roc(val_labels, capsnet_y_pred_prob, n_classes=NUM_CLASSES, 
                        title="CapsNet ROC Curves")
except:
    pass

# Ensemble (meta_model) - if predict_proba is supported for multi-class
try:
    plot_multiclass_roc(val_labels, ensemble_y_pred_prob, n_classes=NUM_CLASSES, 
                        title="Ensemble ROC Curves")
except:
    pass

# 8.5 Feature Map Visualization for ResNet (example of the first conv layer)
#    We just pick 1-2 images from val_imgs and show the filter activations
layer_name = None
for layer in resnet_model.layers:
    # We look for the first convolution or any interesting layer
    if isinstance(layer, layers.Conv2D):
        layer_name = layer.name
        break

if layer_name is not None:
    partial_model = models.Model(inputs=resnet_model.inputs, 
                                outputs=resnet_model.get_layer(layer_name).output)
    
    # Use the first image in val set (just a demo)
    sample_img = val_imgs[0]
    sample_img_input = np.expand_dims(sample_img, axis=0)
    feature_maps = partial_model.predict(sample_img_input)  # shape [1, h, w, channels]
    
    num_channels = feature_maps.shape[-1]
    # Plot a few channels
    plt.figure(figsize=(12, 6))
    for i in range(min(num_channels, 8)):  # show 8 channels
        plt.subplot(2, 4, i+1)
        plt.imshow(feature_maps[0, :, :, i], cmap='viridis')
        plt.axis('off')
    plt.suptitle("Feature Maps from the First Conv Layer (ResNet)")
    plt.tight_layout()
    plt.show()
else:
    print("No Conv2D layer found in ResNet or cannot visualize feature map.")

print("\n=== ALL DONE! ===")
```
